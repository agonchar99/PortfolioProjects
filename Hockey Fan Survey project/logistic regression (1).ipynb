```
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt


df = pd.read_csv('logis_final.csv')


X = df.drop('Q42', axis=1)
y = df['Q42']

X.fillna(X.mean(), inplace=True)
X.fillna(0, inplace=True)

model = LogisticRegression()
model.fit(X, y)


importance = np.abs(model.coef_)[0]

top_10_indices = np.argsort(importance)[-5:]
top_10_features = X.columns[top_10_indices]

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top_10_features, importance[top_10_indices], color='#86bf91')
#ax.set_xlabel('Variable Importance', fontsize=12)
ax.set_ylabel('Features', fontsize=12)
ax.set_title('Top 5 Features Importance', fontsize=16)
ax.grid(axis='x', linestyle='--', alpha=0.8)
#for i, v in enumerate(importance[top_10_indices]):
    #ax.text(v + 0.01, i, str(round(v, 3)), fontsize=10)
plt.show()
```
